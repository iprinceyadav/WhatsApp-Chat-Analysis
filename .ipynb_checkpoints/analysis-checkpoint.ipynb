```python
 import pandas as pd
import re
```

```python
f =open('WhatsApp Chat with ISF-27.txt','r',encoding='utf-8')
```

```python
data=f.read()
```

```python
print(data)
```

```python
pattern = r'\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s?(?:am|pm)?\s-\s'
```

```python
message = re.split(pattern,data)
```

```python
message = [msg.strip() for msg in message if msg.strip()]
```

```python
message
```

```python
dates=re.findall(pattern,data)
dates
```

```python
# Step 1: Fix the narrow spaces



dates_clean = [d.replace('\u202f', ' ').strip() for d in dates]

# Parse the cleaned dates
dates_24h = pd.to_datetime(dates_clean, format='%d/%m/%y, %I:%M %p -')

# Step 3 (Optional): If you want them as strings
dates = dates_24h.strftime('%d/%m/%Y, %H:%M')

# Final output

```

```python
dates
```

```python
df=pd.DataFrame({'user_message':message,'message_data':dates})
#convert message_data type
df['message_data'] = pd.to_datetime(df['message_data'], format='%d/%m/%y, %I:%M %p -', errors='coerce')
df.rename(columns={'message_date':'date'},inplace=True)
df.head()
```

```python
df.shape
```

```python
import re

users = []
messages = []  # Renamed to avoid conflict with the loop variable

# Assuming 'user_message' is your column with user and message data
for entry in df['user_message']:
    # Adjusting the split pattern to match the format more accurately
    result = re.split(r'(\S+):\s', entry)  # Assuming format 'User: Message'
    
    if len(result) > 1:
        users.append(result[1])  # The user name is at position 1
        messages.append(result[2])  # The message content is at position 2
    else:
        users.append('group_notification')  # For group notifications (if applicable)
        messages.append(result[0])  # In case of no user info, just append the message

# Add new columns
df['user'] = users
df['message'] = messages

# Drop the original 'user_message' column
df.drop(columns=['user_message'], inplace=True)
```
